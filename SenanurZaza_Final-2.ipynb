```python
### Senanur Zaza 2022207261

# Q1

#1-a
def getPage(url):
    try:
        import urllib.request
        page = urllib.request.urlopen(url).read()
        page = page.decode("utf-8")
        return page
    except:
        return ""

#1-b
def addToIndex(index, keyword, url):
    if keyword in index:
        # URL zaten varsa sayacı artırmamız, yoksa da yeni url eklediğimiz için count sayısı 1 olması gerekiyor
        if url not in index[keyword]["urls"]:
            index[keyword]["urls"].append(url)
            index[keyword]["count"] += 1
    else:
        index[keyword] = {"urls": [url], "count": 1}

def addPageToIndex(index, url):
    content = getPage(url)
    contentList = content.split()
    for element in contentList:
        addToIndex(index, element, url)
    return index

##Lookup kısmını dictionary indexi için güncelledik
def lookup(index, keyword):
    if keyword in index:
        urls = index[keyword]["urls"]
        count = index[keyword]["count"]
        print("Keyword:", keyword)
        print("Number of results:", count)
        for url in urls:
            print(url)
    else:
        print("Keyword not found.")


def getAllLinks(page):
    urlList = []
    while True:
        url, endpos = getNextTarget(page)
        if url:
            urlList.append(url)
            page = page[endpos:]
        else:
            break
    return urlList

def getNextTarget(page):
    startLink = page.find('<a href=')
    if startLink == -1:
        return None, 0
    startQuote = page.find('"', startLink)
    endQuote = page.find('"', startQuote + 1)
    url = page[startQuote + 1:endQuote]
    return url, endQuote

def crawlWeb(seed):
    tocrawl = [seed]
    crawled = []
    index = {}
    graph = {}
    while tocrawl:
        page = tocrawl.pop()
        if page not in crawled:
            content = getPage(page)
            addPageToIndex(index, page)
            outlinks = getAllLinks(content)
            graph[page] = outlinks
            tocrawl.extend(outlinks)
            crawled.append(page)
    return index, graph

# Testler
index, graph = crawlWeb("https://searchengineplaces.com.tr/")
print("Index:", index)
lookup(index, "of")
lookup(index, "visit")
lookup(index, "historic")
```

```python

def computeRanks(graph):
    d = 0.8
    N = len(graph)
    numloops = 10
    ranks = {}
    for page in graph:
        ranks[page] = 1 / N

    for i in range(numloops):
        newranks = {}
        for page in graph:
            newrank = (1 - d) / N
            for node in graph:
                if page in graph[node]:
                    newrank += d * (ranks[node] / len(graph[node]))
            newranks[page] = newrank
        ranks = newranks

    return ranks

index, graph = crawlWeb("https://searchengineplaces.com.tr/")
ranks = computeRanks(graph)
print(ranks)
```

```python
##Q2

def frequencyOfWord(text):
    ##kelime: kelimenin sıklığı yapısını kullanmak için boş bir dictionary açalım.
    word_frequency = {}
    text = text.lower()

    # Noktalama işaretlerinin yerine boşluk koyalım
    # Eğer noktalama varsa textin içine empty string koyalım, yoksa direkt koyalım.
    cleaned_text = ""
    for char in text:
        if char in '.,;:':
            cleaned_text += ' '
        else:
            cleaned_text += char


    words = cleaned_text.split()
    #print(words)

   ##Aynı kelime her döndüğünde +1 olsun, 1 kere döndüyse 1 kalsın.ç
    for word in words:
        if word in word_frequency:
            word_frequency[word] += 1
        else:
            word_frequency[word] = 1

    for word in word_frequency:
        print('"' + word + '":', word_frequency[word])


frequencyOfWord("In programming , the hard part isn’t solving problems , but deciding what problems to solve.")

frequencyOfWord("selam,:selam:,selam;,SELAM,Selam")

```

```python
#Q3

#3-a

class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height
        self.diagonal_value = (width**2 + height**2) ** 0.5

    def area(self):
        return self.width * self.height
#3-b
    def diagonal(self):
        return self.diagonal_value
#3-c
    def __eq__(self, other):
        return self.width == other.width and self.height == other.height
#3-d
    def __add__(self, other):
        return Rectangle(self.width + other.width, self.height + other.height)
#3-e
    def __str__(self):
        return "Rectangle with width " + str(self.width) + " and height " + str(self.height)

# Testler
rec1 = Rectangle(3, 4)
print(rec1.area())  #  12

print(rec1.diagonal())  #  5

rec2 = Rectangle(3, 4)
rec3 = Rectangle(4, 3)
rec4 = Rectangle(6, 3)

print(rec1 == rec2)  #  True
print(rec2 == rec3)  #  False
print(rec3 == rec4)  #  False

rec5=rec3+rec4
print(rec5.width ,rec5.height) # 10,6

print(rec5) #Output: Rectangle with width 10 and height 6
```
